# GRA Resonance Nullification Experiment
Demonstration: alternating resonant signal averages to zero.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
A = 1.0
gamma = 0.2
T = 1.0         # switching period
duration = 10.0
num_points = 1000

t = np.linspace(0, duration, num_points)
square = np.sign(np.sin(2 * np.pi * t / T))
Xi = A * np.exp(gamma * t) * square

# Moving average over window T
dt = t[1] - t[0]
window = int(T / dt)
cumsum = np.cumsum(np.insert(Xi, 0, 0))
moving_avg = (cumsum[window:] - cumsum[:-window]) / window
t_avg = t[window-1:]

plt.figure(figsize=(10,5))
plt.plot(t, Xi, alpha=0.6, label='Ξ(t)')
plt.plot(t_avg, moving_avg, 'r', linewidth=2, label=f'Moving average (T={T})')
plt.axhline(0, color='black', linestyle='--')
plt.xlabel('Time')
plt.ylabel('Signal')
plt.title('Alternating resonance nullifies in average')
plt.legend()
plt.grid()
plt.show()
```

**Observation**: The moving average converges to zero, confirming GRA-nullification.
